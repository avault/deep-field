# Imports

```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from astropy.io import fits
```

# Load the Data

```python
# Load the fits file
image_file = './data/hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits'
hdu_list = fits.open(image_file)
hdu_list.info()
```

```python
# Select the data
image_data = hdu_list[0].data
```

```python
# Print out some information about the data
print(type(image_data))
print(image_data.shape)
```

# Plot the Data

```python
# Plot the data
fig = plt.figure( figsize=(12,11), facecolor='white' )
ax = plt.gca()

ax.imshow(
    np.log10(image_data), 
    cmap='cubehelix',
    vmin = -4,
)
ax.set_facecolor( 'k' )
ax.set_aspect( 'equal' )
```
